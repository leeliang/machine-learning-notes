# PCA

```python
import matplotlib.pylab as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import numpy as np
%matplotlib inline
```

```python
data = load_iris()
x,y = data.data,data.target
pca = PCA(n_components=2)
z = pca.fit_transform(x)
color = ['red','green','blue']
for i in range(len(set(y))):
    xx = z[np.where(y==i),0]
    yy = z[np.where(y==i),1]
    plt.scatter(xx,yy,color=color[i])
    
```

# NMF

$V_{n*m}=W_{n*k}*H_{k*m}$ $$$$
$W_{n*k}$:特征$$$$
$H_{k*m}$:系数

```python
from sklearn import decomposition
from sklearn.datasets import fetch_olivetti_faces
from numpy.random import RandomState
def plot_gallery(title,images, n_col, n_row):
    plt.figure(figsize=(2. * n_col, 2.26 * n_row)) 
    plt.suptitle(title, size=16)
 
    for i, comp in enumerate(images):
        plt.subplot(n_row, n_col, i + 1)
        vmax = max(comp.max(), -comp.min())
 
        plt.imshow(comp.reshape(image_shape), cmap=plt.cm.gray,
                   interpolation='nearest', vmin=-vmax, vmax=vmax)
        plt.xticks(())
        plt.yticks(())
    plt.subplots_adjust(0.01, 0.05, 0.99, 0.94, 0.04, 0.)
```

```python
n_row,n_col = 2,3
n_components = n_row*n_col
image_shape = (64,64)
dataset = fetch_olivetti_faces(shuffle=True,random_state=RandomState(0))
faces = dataset.data   
faces[0,:].shape
plt.imshow(faces[])
#plot_gallery("First centered Olivetti faces", images[:n_components],n_col,n_row) #取前6幅图像
```

```python
print "faces.shape: %d,%d" % (faces.shape)  
pca = decomposition.PCA(n_components=n_components,whiten=True)
a = pca.fit(faces)
plot_gallery("PCA",a.components_[:n_components],n_col,n_row)
```

```python
nmf = decomposition.NMF(n_components=n_components,init='nndsvda',tol=5e-3)
b = nmf.fit(faces)
plot_gallery("NMF",b.components_[:n_components],n_col,n_row)
```
